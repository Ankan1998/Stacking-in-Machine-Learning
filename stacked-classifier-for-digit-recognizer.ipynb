```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import ExtraTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.ensemble import StackingClassifier
```

```python
ds=pd.read_csv("../input/digit-recognizer/train.csv")
```

```python
ds.head()
```

```python
X=ds.iloc[:,1:].values
y=ds.iloc[:,0].values
```

```python
X_scaled=X/255.
```

```python
estimators = [('knn',KNeighborsClassifier(n_neighbors=10)),
              ('dec_tree',DecisionTreeClassifier(random_state = 2,max_depth = 10)),
              ('extra_tree',ExtraTreeClassifier(random_state = 2,max_depth = 10)),
              ('svm',SVC(gamma = 'auto',random_state = 2)),
              ('random_forest',RandomForestClassifier(n_estimators=10,random_state = 2))                
]

stacked_classif = StackingClassifier(estimators = estimators,final_estimator=RandomForestClassifier(),n_jobs=-1)

stacked_classif.fit(X_scaled,y)
```

```python
ds_test=pd.read_csv("../input/digit-recognizer/test.csv")
```

```python
X_test=ds_test.iloc[:,:].values
```

```python
test_scaled=X_test/255.
```

```python
pred= stacked_classif.predict(X_test)
```

```python
# create submission file
final_submission = pd.DataFrame({'ImageId': np.array(range(1,28001)), 'Label': pred})
final_submission.to_csv("submission.csv", index=False)
```
